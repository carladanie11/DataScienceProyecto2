Analisis y ETL de la Hoja 2: Velocidad_sin_Rangos

```python
# Velocidad_sin_Rangos
df_acc_vel_loc = pd.read_excel(archivo_excel, sheet_name='Velocidad_sin_Rangos')
print("Hoja: Velocidad_sin_Rangos")
print(df_acc_vel_loc.head())
print(df_acc_vel_loc.info())
print(df_acc_vel_loc.describe())
```

```python
# Ruta del archivo Excel
file_path = "https://indicadores.enacom.gob.ar/Files/Datos_Abiertos/Internet.xlsx"  # ruta de tu archivo
# Cargar la hoja específica
df = pd.read_excel(file_path, sheet_name="Velocidad_sin_Rangos")
```

```python
# Opción 2: Eliminar filas con valores nulos
df.dropna(subset=["Velocidad"], inplace=True)
```

```python
# Calcular el percentil 99
p99 = df["Velocidad"].quantile(0.99)

# Filtrar valores anómalos
df = df[df["Velocidad"] <= p99]
```

```python
# Agrupación
df_grouped = df.groupby(["Año", "Trimestre", "Provincia"]).agg(
    Velocidad_Promedio=("Velocidad", "mean"),
    Total_Accesos=("Accesos", "sum")
).reset_index()
```

```python
# Renombrar columnas
df_grouped.rename(columns={
    "Año": "anio",
    "Trimestre": "trimestre",
    "Provincia": "provincia",
    "Velocidad_Promedio": "velocidad_promedio",
    "Total_Accesos": "total_accesos"
}, inplace=True)
```

```python
# Reducir tamaño de las columnas
df_grouped["anio"] = df_grouped["anio"].astype("int16")
df_grouped["trimestre"] = df_grouped["trimestre"].astype("int8")
df_grouped["total_accesos"] = df_grouped["total_accesos"].astype("int32")
```

```python
# Guardar como CSV
df_grouped.to_csv("velocidad_sin_rango_limpio.csv", index=False)
```
